```python
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr)
a=np.zeros((3,3))
print(a)
b=np.ones((3,3))
print(b)
c=np.arange(10)
print(c)
#manuplations
arr_reshaped=arr.reshape(5,1)
print(arr_reshaped)
arr_sliced=arr[2:4]
print(arr_sliced)
```

```python
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
add=(arr1+arr2) #elementary operations
print(add)
print(arr1+3) #broadcasting
```

```python
#vertical
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
stack=np.vstack(arr1+arr2)
print(stack)
```

```python
#horizontal
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
stack=np.stack(arr1+arr2)
print(stack)
```

```python
#split
arr1=np.array([1,2,3,4,5,6])
arr2=np.split(arr1,2)
print(arr2)
```

```python
#transpose
arr1=np.array([[1,2,3],[4,5,6]])
arr2=arr1.T
print(arr1)
print(arr2)
```

```python
#linear algebra using numpy
arr1=np.array([[1,2],[5,6]])
arr2=np.array([[2,3],[4,5]])
arr3=np.dot(arr1,arr2)
d=np.linalg.eig(arr3)
print(d)
```

```python
import numpy as np
arr1=np.array([[1,2,3],[4,5,6,]])
arr2=np.array([[1,2,3],[4,5,6,]])
arr3=np.sum(arr1,axis=0)#column wise
print(arr3)
arr4=np.sum(arr1,axis=1)#row wise
print(arr4)
```

```python
a=np.array([[2,3,4],[1,3,5]])
c=np.sum(a, axis=0)
d=np.sum(a, axis=1)
print(a)
```

```python
#statistical
arr1=np.array([1,2,3,4,5,6])
arr2=np.array([[1,2,3],[4,5,6,]])
print("mean",np.mean(arr1))
print("median",np.median(arr1))
print("variance",np.var(arr1))
print("standard deviation",np.std(arr1))
```

```python
import matplotlib.pyplot as plt
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]
profit_a=[(a[i]-a[i-1])for i in range(1,len(a))]
profit_b=[(b[i]-b[i-1])for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='pink',linewidth='2',label='Company',marker='*',ms='15',mec='r')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label='Company',ms='10',marker='>')
plt.show()
```
